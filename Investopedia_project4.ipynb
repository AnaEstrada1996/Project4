### Analyzing Working Capital 

I used this page for some help with understanding what working capital was: https://www.investopedia.com/terms/w/workingcapital.asp

##### What is working capital?
Working capital is difference between a company's current assets and its current liabilities. It can help to measure a comapanies' short-terrm financial health. 
A company with a negative working capital means that the ratio of their current assets to liabilities are less than one. 
A positive working capital means that the company can fund its current operations and can invest in future growth. 
A high working capital does not always mean that it is good. It could mean that the business has too much inventory or it is not investing its excess cash.

Next I will be calculating the drift. The drift is the direction that rates of returns have had in the past. 

```python
returns = data.pct_change()
returns.dropna(inplace=True)
returns.head()
```

Next I will be working with variance eand daily returns which can help with the voltality of my shares

```python
def mont_car(symbol):
    days = 7
    trials = 20

    mean = returns.mean()[symbol]
    var = returns.var()[symbol]
    std = returns.std()[symbol]
    drift = mean - (0.5*var)

    
    Z = norm.ppf(np.random.rand(days, trials))
    daily_returns = np.exp(drift + std * Z)
    

    price_paths = np.zeros_like(daily_returns)
    price_paths[0] = data[symbol].iloc[-1]
    for t in range(1, days):
        price_paths[t] = price_paths[t-1]*daily_returns[t]
    mplt.plot(price_paths)
```

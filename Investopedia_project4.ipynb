## Looking at my companies' risk
I will be chcking the standard deviation and mean for my companies. My goal is to see whether the standard deviation is larger than the mean, if it is, then my stocks can exhibit a higher risk. 
I used this: https://www.investopedia.com/terms/s/standarddeviation.asp in order to better understand the importance of standard deviation compared to mean

```python
#Here I am getting the mean
comp_mean = returns.mean()
#Here I am getting the variance
comp_var = returns.var()
#Here I am getting the stnd dev
comp_stdev = returns.std()
comp_stdev2 = pd.DataFrame(returns.std())


#Here I am creating a dataframe for mean and for stdev
mean_df = pd.DataFrame(comp_mean)
mean_df2 = mean_df.rename(columns={0: 'mean'})

stdev_df = pd.DataFrame(comp_stdev)
stdev_df2 = stdev_df.rename(columns={0: 'std'})

print(stdev_df2.head())
print(mean_df2.head())
```

##### What I gathered from the stock returns and stdev and mean

Volatility: The rate at which the price of a stock increases or decreases over a period of time 

Standard deviation helps measure the security volatility of stocks. A stock with a greater standard deviation compared to mean can be that it is a volatile stock which a lower standard deviation can mean that the stock is not volatile. A stock that has a low standard deviation is not preferred. 

In conclusion, my standard deviation is greater than the means so I can say that my stocks are highly volatile. Because they are volatile, I have the potential to make more money quickly but it does impose a greater risk.
